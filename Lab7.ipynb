```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pickle
```

```python
infile1 = open ("higgs_100000_pt_250_500.pkl",'rb')
dict_higgs = pickle.load(infile1)
infile2 = open ("qcd_100000_pt_250_500.pkl",'rb')
dict_qcd = pickle.load(infile2)
```

# 1

```python
plt.title('Mass Variable')
plt.xlabel('Mass')
plt.ylabel('Counts')
plt.hist([dict_higgs['mass'],dict_qcd['mass']],bins=800,density=False,alpha=1,label=['Higgs','QCD'],stacked=True);
plt.xlim(0,250)
plt.ylim(0,1400)
plt.legend()
plt.show()
```
